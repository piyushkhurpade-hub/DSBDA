```python
df['salary'] = df['salary'].fillna(df['salary'].mean())
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])
```

```python
df['status'] = le.fit_transform(df['status'])
```

```python
df['workex'] = le.fit_transform(df['workex'])
```

```python
df['specialisation'] = le.fit_transform(df['specialisation'])
df.head()
```

```python
df.boxplot(column=['ssc_p','hsc_p','degree_p','mba_p'])
plt.title("Boxplot of Academic Scores")
plt.show()
```

```python
z = np.abs(stats.zscore(df['mba_p']))
df[z > 3]
```

```python
Q1 = df['hsc_p'].quantile(0.25)
Q3 = df['hsc_p'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
print("Lower Bound:", lower)
print("Upper Bound:", upper)
```

```python
df[(df['hsc_p'] < lower) | (df['hsc_p'] > upper)]
```

```python
median = df['hsc_p'].median()
df['hsc_p'] = np.where((df['hsc_p'] < lower) | (df['hsc_p'] > upper),median,df['hsc_p'])
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["salary"], kde=True, bins=30, color="blue")
plt.title("Salary Before Log")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()
```

```python
df["log_salary"] = np.log(df["salary"])
plt.figure(figsize=(8,5))
sns.histplot(df["log_salary"], kde=True, bins=30, color="green")
plt.title("Salary After Log")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.scatter(df['mba_p'], df['salary'])
plt.xlabel("MBA Percentage")
plt.ylabel("Salary")
plt.title("MBA vs Salary")
plt.show()
```
