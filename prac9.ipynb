```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=sns.load_dataset('titanic')
```

```python
dataset.isnull().sum()
```

```python
dataset['age'] = dataset['age'].fillna(dataset['age'].mean())
```

```python
dataset.isnull().sum()
```

```python
sns.boxplot(x='sex',y='age',hue = 'survived',data=dataset)
plt.title('Age Distribution of Survived and Unsurvived Data')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.show()
```

Female passengers show a noticeably higher survival rate compared to male passengers.
Male non‑survivors dominate the middle‑age range, while male survivors are generally younger.Children (low age values) have better survival chances across both sexes.The median age of male survivors is lower than that of male non‑survivors.Female survivors display a wider age spread compared to female non‑survivors.Outliers include very old passengers in both survived and non‑survived groups.The interquartile range for female survivors is larger than for female non‑survivors.Overall, survival patterns differ significantly between male and female age distributions.
